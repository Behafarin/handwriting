```python
import pandas as pd
import numpy as np
from sklearn import svm
```

```python
semeion = pd.read_csv('semeion.data', delimiter=r"\s+")
semeion
```

```python
def frac(df, fraction):
    return df.sample(frac=fraction)
df = frac(semeion, 0.8)
learning_data = df.iloc[:, 0:255]
learning_labels = df.iloc[:,256:266]
test = semeion.merge(learning_data, how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']
test_data = test.iloc[:,:255]
test_labels = test.iloc[:,256:266]
labels = []
ans = []
labels = learning_labels.to_string(header=False,
                  index=False,
                  index_names=False).split('\n')
ans = test_labels.to_string(header=False,
                  index=False,
                  index_names=False).split('\n')
```

learn here

```python
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(learning_data, labels)
```

```python
my_result = clf.predict(test_data)
```

```python
counter = 0
intersection = []
for i in range(318):
    if my_result[i] == ans[i]:
        intersection.append(ans[i])
        counter += 1
print((counter/318)*100)
```
